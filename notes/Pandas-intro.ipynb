```python
# join and merge
imports_exports_DoL = {
    'imports': [457, 643, 2342],
    'exports': [418, 441, 1545]
}

codes = pd.Index(['CAN','GBR','USA'] , name='code')

join1 = pd.DataFrame(imports_exports_DoL, index=codes)
```

```python
country_landDoL = {
    'country': ['Belgium','United Kingdom','United States','Vietnam'],
    'land': [30000, 240000, 91000, 300000]
}
codes = pd.Index(['BEL','GBR','USA','VNM'])

join2 = pd.DataFrame(country_landDoL,index = codes)
```

```python
join1
```

```python
join2
```

```python
# join operates on index!
# left join
join1.join(join2,how='left')
```

```python
# inner join
join1.join(join2,how='inner')
```

```python
join1.join(join2,how='right')
```

```python
# equivalent to the cell above, note the difference in column order
join2.join(join1, how='left')
```

```python
# merge() can use any column in a dataframe
# the following example uses the only common column between the dfs, year.
ind_table = ind[ind.country=="United States"].reset_index(drop=True)
ind_table = ind_table[['year','pop']]
```

```python
topnames = pd.read_csv('https://raw.githubusercontent.com/ernbilen/Data200_Fall22/main/data/topnames.csv')
```

```python
ind_table.head()
```

```python
topnames.head()
```

```python
# merging the two dataframes together
pd.merge(ind_table,topnames,how='right', on='year')
```

```python
# equivalent to the cell above
ind_table.merge(topnames,on='year',how='right')
```

```python
# melt and pivot functions
# melt() converts df from wide to long
names_by_year = pd.DataFrame({
    'sex':['Female','Male'],
    2015: ['Emma','Noah'],
    2016: ['Emma','Noah'],
    2017: ['Emma','Liam']
})
names_by_year
```

```python
# convert to long format using melt()
names_by_year.melt(id_vars = 'sex', var_name = 'year', value_name = 'name')
```
